```python
import pickle

import matplotlib.pyplot as plt

import numpy as np
```

```python
p = pickle.load(open('../params_MT_654.pkl', "rb"))
```

```python
for k, v in p.items():
    plt.figure(figsize=(10, 10))
    
    if type(v) is np.ndarray and len(v) > 0:
        plt.hist(v, bins=30)
        plt.title(k)
        
    else:
        print(k, v)
        
    plt.show()
```
